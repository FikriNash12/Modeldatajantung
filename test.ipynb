```python
# Import Library
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Membaca file csv dan menampilkan 5 data teratas
df = pd.read_csv('japan_heart_attack_dataset.csv')
df.head()
```

```python
# Menghitung kolom dan baris dari dataset
df.shape
```

```python
# Menampilkan info dari dataset
df.info()
```

```python
# Menampilkan deskripsi dari dataset
df.describe()
```

```python
# Menampilkan deskripsi data dengan tipe data Object
df.select_dtypes(include='object').describe()
```

```python
# Mencari nilai NaN dalam dataset
df.isnull().sum().sort_values(ascending=False)
```

```python
# Mengatasi nilai NaN dengan nilai yang paling sering muncul
mode_value = df['Alcohol_Consumption'].mode()[0]
df['Alcohol_Consumption'] = df['Alcohol_Consumption'].fillna(mode_value)
```

```python
# Mencari nilai duplikat
df.duplicated().sum()
```

```python
# Cek nilai NaN sudah terhapus atau belum
data = df.isnull().sum()
data
```

```python
# 3. Univariate Analysis
# Numerik
numerical_cols = ['Age', 'Cholesterol_Level', 'BMI', 'Heart_Rate', 'Systolic_BP', 'Diastolic_BP', 'Stress_Levels']
for col in numerical_cols:
    plt.figure(figsize=(8, 4))
    sns.histplot(df[col], kde=True, bins=30)
    plt.title(f'Distribution of {col}')
    plt.show()
```

```python
# Kategorikal
categorical_cols = ['Gender', 'Region', 'Smoking_History', 'Diabetes_History', 'Hypertension_History', 
                    'Physical_Activity', 'Diet_Quality', 'Alcohol_Consumption', 'Family_History', 'Heart_Attack_Occurrence']
for col in categorical_cols:
    plt.figure(figsize=(8, 4))
    sns.countplot(x=col, data=df)
    plt.title(f'Count Plot of {col}')
    plt.show()
```

```python
# 4. Bivariate Analysis
# Numerik
for col in numerical_cols:
    plt.figure(figsize=(8, 4))
    sns.boxplot(x='Heart_Attack_Occurrence', y=col, data=df)
    plt.title(f'{col} vs Heart Attack Occurrence')
    plt.show()
```

```python
# Kategorikal
for col in categorical_cols:
    if col != 'Heart_Attack_Occurrence':
        plt.figure(figsize=(8, 4))
        sns.countplot(x=col, hue='Heart_Attack_Occurrence', data=df)
        plt.title(f'{col} vs Heart Attack Occurrence')
        plt.show()
```

```python
# 5. Multivariate Analysis
# Korelasi
plt.figure(figsize=(10, 8))
sns.heatmap(df[numerical_cols].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
# 6. Feature Engineering (Example: Creating Age Groups)
df['Age_Group'] = pd.cut(df['Age'], bins=[0, 30, 50, 70, 100], labels=['0-30', '31-50', '51-70', '71-100'])
```
